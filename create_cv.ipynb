```python
#Unzip

from pathlib import Path
import gzip
import shutil

input_dir = Path("2010-05")
output_dir = Path("2010-05-uncompressed")
output_dir.mkdir(parents=True, exist_ok=True)

for gz_path in sorted(input_dir.glob("*.json.gz")):
    out_path = output_dir / gz_path.name.replace(".json.gz", ".json")
    with gzip.open(gz_path, "rb") as f_in, open(out_path, "wb") as f_out:
        shutil.copyfileobj(f_in, f_out)

print(f"Extracted to: {output_dir.resolve()}")
```

```python
#Convert to CSV format, Twitter data

from pathlib import Path
import json
import csv

input_dir = Path("2010-05-uncompressed")
output_dir = Path("2010-05-csv")
output_dir.mkdir(parents=True, exist_ok=True)

for json_path in sorted(input_dir.glob("*.json")):
    csv_path = output_dir / (json_path.stem + ".csv")
    with csv_path.open("w", newline="", encoding="utf-8") as f_out:
        writer = csv.writer(f_out)
        writer.writerow(["Text", "Origin", "id"])

        with json_path.open("r", encoding="utf-8") as f_in:
            for line in f_in:
                line = line.strip()
                if not line:
                    continue
                try:
                    record = json.loads(line)
                except json.JSONDecodeError:
                    continue

                text = record.get("text", "")
                msg_id = record.get("id") or record.get("id_str") or ""
                writer.writerow([text, "Twitter", msg_id])

print(f"Wrote CSVs to: {output_dir.resolve()}")

```
