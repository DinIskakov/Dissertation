```python
import pandas as pd

# df = pd.read_csv("nyt.csv")  # if needed
df["pub_date"] = pd.to_datetime(df["pub_date"], errors="coerce")

matches = df[(df["pub_date"].dt.year == 2011) & (df["pub_date"].dt.month == 12)]

out = pd.DataFrame({
    "Text": matches["lead_paragraph"]
        .fillna(matches["abstract"])
        .fillna(matches["snippet"])
        .fillna(""),
    "Origin": "NYTimes",
    "id": matches["_id"].fillna(""),
    "country": "United States of America",
    "language": "en",
})

out.to_csv("nyt_2011_12.csv", index=False)
```
